### Import necessary packages 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress Pandas Future Warning.
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### Import Preprocessed Data

```python
ratings_distribution = pd.read_csv("./data/ratings_distribution.csv")
rating_count_sorted = pd.read_csv("./data/rating_count_sorted.csv")
user_gender_distribution = pd.read_csv("./data/user_gender_distribution.csv")
user_occupation_distribution = pd.read_csv("./data/user_occupation_distribution.csv")
```

## Visualization

### Distribution of the total counts per each rating

```python
sns.barplot(
    x = 'rating',
    y = 'count',
    data = ratings_distribution,
    palette = sns.color_palette('pastel', 5)
).set_title('Distribution of the total count per each rating')
```

### Distribution of each movie's total rating count

```python
# movie ratings are heavily skewed
plt.plot(
    rating_count_sorted['rating']
)
plt.title("Distribution of each movie's total rating count")
```

### Distribution of user's gender

```python
# Highly skewed towards male users
sns.barplot(
    x = 'gender',
    y = 'count',
    data = user_gender_distribution
).set_title("Distribution of user's gender")
```

### Distribution of user's occupation

```python
plt.figure(figsize=(12, 8))

ax = sns.barplot(
    x = 'occupation',
    y = 'count',
    data = user_occupation_distribution,
    palette=sns.color_palette('pastel' , len(user_occupation_distribution))
)

ax.set_xticklabels(ax.get_xticklabels(), rotation=70)
ax.set_title("Distribution of user's occupation")

plt.show()
```
